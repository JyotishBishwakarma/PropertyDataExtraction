```python
import pandas as pd
```

```python
raw_df = pd.read_csv('Palm_beach_final_raw_data/rawData_pg_102-104_FL_palm_beach_county.csv')
final_df = pd.read_csv('FinalData/FL_PalmBeachCounty.csv')

raw_df.head()
```

```python
#function to extract city, state and zip code
def extract_address_info(address):
    try:
        parts = address.split()
        if len(parts) >= 3:
            city = ' '.join(parts[:-3])
            state = parts[-3]
            zip_code = parts[-2]
            return city, zip_code, state,
        else:
            return None, None, None
    except Exception as e:
            return None, None, None 
```

```python
# Display the DataFrame
#print(df.head())
raw_df[['CITY', 'Zipcode', 'State' ]] = raw_df['Address'].apply(extract_address_info).apply(pd.Series)

# Drop the original 'Address' column if needed
raw_df = raw_df.drop('Address', axis=1)

print(raw_df.head())
```

```python
raw_df = raw_df[['ParcelId', 'Owner Name', 'Property Address', 'CITY','County','Zipcode','State','Property Value','Land Use']]
```

```python
raw_df.head()
```

```python
df_result = pd.concat([final_df, raw_df], ignore_index=True)

df_result.to_csv('FinalData/FL_PalmBeachCounty.csv',index=False)
```
